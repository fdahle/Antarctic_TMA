```python
import os
import Metashape

project_fld = "/data/ATM/data_1/sfm/agi_projects/"
project_name = "agi_agi"

# create the path to the project
project_path = os.path.join(project_fld, project_name, project_name + ".psx")

doc = Metashape.Document(read_only=False)  # noqa
doc.open(project_path, ignore_lock=True)
chunk = doc.chunk
```

```python
import src.load.load_image as li

# get the camera names
camera_names = [camera.label for camera in chunk.cameras]

images = []
for img_name in camera_names:
    img = li.load_image(img_name)
    images.append(img)
```

```python
import pandas as pd

csv_pth = os.path.join(project_fld, project_name, "debug", "tie_points.csv")

# read the csv
df = pd.read_csv(csv_pth)

print(df)
```

```python
# order the images by 'tp_id'
df = df.sort_values(by=['keypoint_left'])

# iterate over all rows
for index, row in df.iterrows():
    
    left_img = images[int(row["image_id_left"])]
    right_img = images[int(row["image_id_right"])]
    
    # get the pixel value of tie point for left image
    left_px_val = left_img[int(row["y_left"]), int(row["x_left"])]
    
    # get the pixel value of tie point for right image
    right_px_val = right_img[int(row["y_right"]), int(row["x_right"])]
    
    # get the average pixel of the tie point and a 3x3 window around it
    #left_px_avg = left_img[int(row["y_left"])-1:int(row["y_left"])+1, int(row["x_left"])-1:int(row["x_left"])+1].mean()
    #right_px_avg = right_img[int(row["y_right"])-1:int(row["y_right"])+1, int(row["x_right"])-1:int(row["x_right"])+1].mean()
    
    pixel_avg = int((int(left_px_val) + int(right_px_val)) / 2)
    
    print(row['track_id'], pixel_avg, left_px_val, right_px_val)
```
